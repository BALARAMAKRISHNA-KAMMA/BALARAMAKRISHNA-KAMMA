```python
def avg_word(k):
    words =k.split()
    return sum(len(word) for word in words)  / len(words)
```

```python
dframe['average_word_length']= dframe['body'].apply(lambda x : avg_word(x))
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
print(stop_words)
```

```python
len(stop_words)
```

```python
dframe['stop_word_count']= dframe['body'].apply(lambda x : len([word for word in x.split() if word.lower() in stop_words]))
```

```python
dframe
```

```python
dframe['lower_case']= dframe['body'].apply(lambda x : ' '.join(word.lower() for word in x.split()))
```

```python
dframe
```

```python
dframe['punct']=dframe['lower_case'].str.replace('[^\w\s]','')
```

```python
dframe['stop_word']=dframe['punct'].apply(lambda x:' '.join(word for word in x.split() if word not in stop_words))
```

```python
' '.join(dframe['stop_word']).split()
```

```python
pds.Series(' '.join(dframe['stop_word']).split()).value_counts()[:50]
```

```python
other_s_w = ['phone','battery','screen','one','get','camera','phones','works','samsung','time','work','back','also','got','sim','iphone','day']
```

```python
dframe['clean_sentenses'] = dframe['stop_word'].apply(lambda x:' '.join(word for word in x.split() if word not in other_s_w))
```

```python
dframe
```

```python
dframe['clean_sentenses_word_count'] = dframe['clean_sentenses'].apply(lambda x : len(x.split()))
```

```python
dframe['clean_sentenses_rate'] = dframe['clean_sentenses_word_count'] / dframe['word_count'] 
```

```python
dframe['limitizE'] = dframe['clean_sentenses'].apply(lambda x :" ".join(Word(word).lemmatize() for word in x.split()))
```

```python
from textblob import TextBlob
```

```python
dframe['limitizE'].apply(lambda x : TextBlob(x).sentiment)
```

```python
dframe['polarity'] = dframe['limitizE'].apply(lambda x : TextBlob(x).sentiment[0])
```

```python
dframe.loc[dframe['polarity'] > 0, 'Sentiment'] = 'Positive'
```

```python
dframe.loc[dframe['polarity'] == 0, 'Sentiment'] = 'Neutral' 
```

```python
dframe.loc[dframe['polarity'] < 0, 'Sentiment'] = 'Negative'
```
